```python
import numpy as np
import pandas as pd
import csv
from datetime import datetime
```

```python
def createDF(infile_name):
    DF = pd.read_csv(infile_name, header=0, sep=',')
    return DF
    
```

```python
#read data tables into pandas
nfl_stats_2017 = createDF("nflstats2017.csv")
timezones = createDF("timezones2017.csv")

#create table for away team zones to join
Away_timezones = timezones[['Team_Code','Team_Zone','Zone_Value']]
Away_timezones = Away_timezones.rename(index=str, columns = {"Team_Code":"Away","Team_Zone":"AZone","Zone_Value":"AZoneVal"})

#create table for home team zones to join
Home_timezones = timezones[['Team_Code','Team_Zone','Zone_Value']]
Home_timezones = Home_timezones.rename(index=str, columns = {"Team_Code":"Home","Team_Zone":"HZone","Zone_Value":"HZoneVal"})

#Join timezone info to main dataframe
nfl_stats_2017 = nfl_stats_2017.join(Away_timezones.set_index('Away'), on='Away',how = 'left')
nfl_stats_2017 = nfl_stats_2017.join(Home_timezones.set_index('Home'), on='Home',how = 'left')

#create time zone diff for away team
nfl_stats_2017 = nfl_stats_2017.assign(AZDiff = lambda x: nfl_stats_2017.HZoneVal - nfl_stats_2017.AZoneVal)

nfl_stats_2017
```
